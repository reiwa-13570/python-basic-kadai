```python
# 上辺：10cm
# 下辺：20cm
# 高さ：5cm

# 公式：台形の面積 =（上辺＋下辺）× 高さ ÷ 2

上辺 = 10
下辺 = 20
高さ = 5

print(f"面積は {(上辺 + 下辺) * 高さ / 2} ㎠です。")
```
